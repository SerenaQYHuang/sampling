### 1. Random sampling according to rows and columns

```python
import pandas as pd 
df = pd.read_csv('openrice_sample.csv',header = None) 
#Set the header as none so that the columns will be numbered, making sampling possible
df
```

### I need 100 random samples from all the cells in the csv for my research project (another dataset)
### First, I tried sampling 50 rows and 2 columns at the same time

```python
import numpy as np
i=np.random.randint(0,931,50) # Sample 50 from 932 rows
j=np.random.randint(0,8,2) # Sample 2 from 8 columns
```

```python
newdf = df.loc[i,j] # loc the rows and columns sampled
newdf
```

### However, this was not what I expected. 
### I wanted a random 100 samples from the 932 X 9 cells, not two columns.
### So I tried another way as follows
### First, I created a list of the cells I wanted

```python
rows = []
cols = []
for a in range(100):
# Creating a for loop to add cells into the list
    i = np.random.randint(0,931)
    j = np.random.randint(0,8)
    rows.append(i)
    cols.append(j)
    a+=1
```

```python
dfdict = {}
# Creating a dictionary for the cells sampled
for a in range(100):
    i = rows[a]
    j = cols[a]
    dfdict[i+1,j+1] = df.values[i][j] 
# the keys should be i+1, j+1 because the rows and columns in dataframes start with 0, while in csv they start with 1
    a+=1
```
